## Seaborn LINE PLOT

```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Generating DATA
#Seaborn takes a PANDAS DATAFRAME to generate plot.
x = np.random.randint(1,10,(100))
y = np.random.randint(1,10,(100))

data = pd.DataFrame({"days":x,
                    "temperature":y})
```

```python
sns.lineplot(x = "days",y = "temperature", data=data)
plt.show()
```

```python
#Import Real Dataset from Seaborn
tips_data = sns.load_dataset("tips", cache = True)
```

```python
tips_data.info()
```

```python
tips_data.shape
```

```python
sns.lineplot(x="total_bill", y="tip", data = tips_data)
plt.show()
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

sns.lineplot(x="size",
             y="tip",
             data = tips_data,
             marker= "D",
             hue="sex", legend= "full")
plt.xlabel("SIZE")
plt.ylabel("T I P")
plt.title("Line PLOT Seaborn")
plt.show()
```

## Get Background GRID

```python
plt.figure(figsize=(16,9))
sns.set(style="darkgrid")
lineplot = sns.lineplot(x="size",
             y="tip",
             data = tips_data,
             marker= "D",
             hue="sex", legend= "full")
plt.xlabel("SIZE")
plt.ylabel("T I P")
plt.title("Line PLOT Seaborn")
plt.show()
```

## Save Figure

```python
lineplot.figure.savefig("Sea-Line.png")
```

## Seaborn HISTOGRAM / DISTPLOT

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

sns.displot(tips_data["size"])
plt.show()
```

```python
sns.histplot(tips_data["size"])
plt.show()
```

```python
sns.displot(tips_data["size"], kde=True, #kernel density estimator
            color = "k",
            alpha = 0.5
           )
plt.show()
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.


# LABELLING THE AXES


sns.displot(tips_data["size"], kde=True, #kernel density estimator
            color = "k",
            alpha = 0.5,
            label = "X Label"
           )

plt.title("Histogram", fontsize=15)
plt.legend()
plt.show()
```

```python
tips_data.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(16,9))
sns.set()
sns.displot(tips_data["total_bill"], kde=True, #kernel density estimator
            color = "k",
            alpha = 0.5,
            label = "X Label",
            bins = bins
           )
plt.legend()
plt.xticks(bins)
plt.show()
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

#Rug PLOT

plt.figure(figsize=(16,9))
sns.set()
sns.displot(tips_data["total_bill"], kde=True, #kernel density estimator
            color = "k",
            alpha = 0.5,
            label = "X Label",
            rug = True
            
        
           )
plt.legend()
plt.xticks(bins)
plt.show()
```

## Seaborn Barplot / BarGraph

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

order = ['Sun', 'Thur', 'Fri', 'Sat' ]
hue_order = ['Female', 'Male']



sns.barplot(x = "day",
            y = "total_bill",
            hue = "sex",
            data = tips_data,
            order = order,
            hue_order = hue_order,
            estimator = np.max,
            ci = 50 #black line length in between bars.
            
           )


plt.legend(loc=2)
plt.show()
```

```python
data = pd.read_csv("Global Warming.csv")
```

```python
data.info()
```

```python
# Drop Columns because HEATMAP only works with numeric values.

remove_col = ["Country Code", "Indicator Name", "Indicator Code"]

short_data = data.drop(remove_col, axis=1).set_index("Country Name")
```

```python
# SET X, Y Label

short_data.head(6)
plt.figure(figsize=(16,9))
ax = sns.heatmap(short_data,
            annot = True,  #Shows Values on the HEATMAP
           )

ax.set(
        xlabel="CO2 Emissions Per Year",
        ylabel="Country Name",
)

plt.show()
```

## SEABORN Correlation

```python
corr = short_data.corr()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(corr, annot=True, linewidth = 3)
plt.show()
```

```python
#Convert Data into PANDAS DATAFRAME.

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

data  = pd.DataFrame(np.c_[data['data'], data['target']],
                    columns = np.append(data['feature_names'],['target']))
```

```python
corr = data.corr()
```

```python
plt.figure(figsize=(25,25))
ax = sns.heatmap(corr, annot=True, linewidth = 3)
plt.title("Heatmap of Breast Cancer")
plt.show()
```
