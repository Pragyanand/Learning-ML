## Import Matplotlib Library

```python
import matplotlib.pyplot as plt
import random
```

```python
#Generating Random Data to Plot

x = random.sample(range(1,100),5)
y = random.sample(range(1,100),5)
z = random.sample(range(1,100),5)
```

## Plotting the Data - LINE

```python
# This plots a simple line plot.

plt.plot(x,y)

#Adding a Title
plt.title("Random Number Vs Random Number")

#Adding the X-axis & Y-axis Labels

plt.xlabel("R1")
plt.ylabel("R2")

plt.show()

```

## Customizing the Plot

```python
#Plotting the Data.
# Markers can be added here.
# Colors can also be changed.
# Change the linetype, linewidth.

# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

plt.plot(x,y, color="k", marker="D", linestyle=":", linewidth=2,
        markersize=3)

#Adding a Title
plt.title("Random Number Vs Random Number")

#Adding the X-axis & Y-axis Labels

plt.xlabel("R1")
plt.ylabel("R2")

plt.show()
```

## Shortening the Customization 

```python
# Color, Marker and linetype can be changed in one argument only!

# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

plt.plot(x,y, "kD:", linewidth=2, markersize=3)

# Adding a Title
# Change the font size.
plt.title("Random Number Vs Random Number", fontsize=15)

#Adding the X-axis & Y-axis Labels

plt.xlabel("R1", fontsize=15)
plt.ylabel("R2", fontsize=15)

plt.show()
```

## Creating the Legend

```python
# Color, Marker and linetype can be changed in one argument only!

# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

plt.plot(x,y, "kD:", linewidth=2, markersize=3)

# Adding a Title
# Change the font size.
plt.title("Random Number Vs Random Number", fontsize=15)

#Adding the X-axis & Y-axis Labels

plt.xlabel("R1", fontsize=15)
plt.ylabel("R2", fontsize=15)
plt.legend(["This is legend"],loc=4)

# You can also use label in plt.plot for
# labelling the line
# Location of the legend can be set with 'loc' with
# values ranging from 0-4.
plt.show()
```

## Get the Background Grid

```python
#import style module for the grid

from matplotlib import style

style.use('ggplot')
plt.plot(x,y, 'ro:')

plt.grid(color="k", linestyle="-", linewidth = 1)

plt.show()
```

## Plotting Multiple Lines

```python
plt.plot(x,y,'yo:')
plt.plot(z,y,'mD--')

plt.title("This Graph is Useless")

plt.xlabel( "This is X Label", fontsize = "13" )
plt.ylabel( "This is Y Label", fontsize = "13" )

plt.show()
```

## Plotting the Data - HISTOGRAM

```python
import numpy as np

men = np.random.randint(20,50,(100))
women = np.random.randint(15,63,(100))
```

```python
print("Men : ", men,"\n")
print("Women : ", women)
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

# If str object not callable error gets in your way
# then restart the kernel!

plt.hist(men)
plt.title("Men Histogram")
plt.xlabel("Men X Label")
plt.ylabel("Men Y Label")
plt.show()
```

## Formatting the Histogram

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

# BINS is the thickness/category of each bar.

bins = [20,25,30,35,40,45,50]

# Rwidth is width of each bar, values rango from 0 to 1.
plt.hist(men, bins,rwidth=0.8, histtype="bar", color="k")
plt.legend(["LABEL OF HIST"], loc=2)
```

## Plotting Multiple Histogram Together

```python
# Plotting Histogram differently results in this which doesnt provide 
# much insight into the data.
plt.hist(men, bins,rwidth=0.8, histtype="bar", color="k")
plt.hist(women, bins,rwidth=0.8, histtype="bar", color="r")
plt.legend(["LABEL OF Men", "Label of Women"], loc=2)
plt.show()
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

# Pass all parameters in a list in the single
# plt.hist() function.

plt.figure(figsize = (16,9)) #Change size of Histogram
plt.hist([men,women], bins,rwidth=0.8, histtype="bar", color=["k","r"], label=["Men","Women"])
plt.legend(loc=2)

plt.show()
```

## Plotting Bar Chart

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
# Vertical Bar Chart
plt.bar(classes, class1_students, color = "k")
```

```python
# Horizaontal Bar Chart
plt.barh(classes, class1_students, color = "k")
```

```python
#Change width of BAR
plt.bar(classes, class1_students, color = "k", width=0.2)
```

```python
#Align the Bar
plt.bar(classes, class1_students, color = "k", width=0.2, align = "edge")
```

```python
#Change bar border color

plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor="g")
```

```python
#Change border width 
plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor = "r"
       ,linewidth = 2)
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

#Change border style, linewidth
plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor = "r"
       ,linewidth = 2, linestyle="--")
```

```python
#Hide / Show graph
plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor = "r"
       ,linewidth = 2, linestyle="--", visible = False)
```

```python
#Change opacity of BARs.
plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor = "r"
       ,linewidth = 2, linestyle="--", alpha = 0.5)
```

```python
#Setting title, xlabel and ylabel
plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor = "r"
       ,linewidth = 2, linestyle="--")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title of BAR Chart")
```

```python
#Plotting multiple arrays
#Hard to get any insight.

plt.bar(classes, class1_students, color = "k", width=0.2, edgecolor = "r"
       ,linewidth = 2, linestyle="--")
plt.bar(classes, class2_students, color = "b", width=0.2, edgecolor = "r"
       ,linewidth = 2, linestyle="--")
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

#Plotting Multiple arrays, RIGHT WAY!


classes_index = np.arange(len(classes))
print("Classes Index : ",  classes_index)
width = 0.2

plt.bar(classes_index, class1_students, width, color='b',
        label = "Class 1 Students")

plt.bar(classes_index + width, class2_students, width, color='k',
        label = "Class 2 Students")

plt.bar(classes_index + 2*width, class3_students, width, color='y',
        label = "Class 3 Students")

plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title of BAR Chart")
plt.legend()

#Xticks shows the XLABEL, when plotting multiple arrays.
plt.xticks(classes_index + width, classes, rotation = 20)
plt.show()
```

## Plotting SCATTER PLOT

```python
import pandas as pd

reviews = pd.read_csv("googleplaystore.csv" ,nrows = 1000)
```

```python
reviews.info()
```

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.

plt.figure(figsize = (16,9))
plt.scatter(reviews["Rating"], reviews["Reviews"],c = "b" #c for color
            , marker="*"
           ,s=100 #s is for markersize
           , alpha = 0.8 #marker transparency
            ,linewidths = 5 #marker width
           )

plt.scatter(reviews["Rating"], reviews["Installs"],c = "r" #c for color
            , marker="*"
           ,s=100 #s is for markersize
           , alpha = 0.8 #marker transparency
            ,linewidths = 5 #marker width
           )

plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.title("Rating vs Reviews")
plt.show()
```

## Slicing PIE Chart

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.



#We'll use EXPLODE for that.

explode = [0.03,0,0.1,0.3,0]  # Write zeros equal to number of classes.

# set zeros to non-zero value for ones you want to
# slice them off the pie.

# Greater value = Greater Distance from the pie. 

color = ['c', 'r', 'g', 'y', 'k']
textprops = {"fontsize":11}

plt.pie(class1_students,
        labels = classes,
        explode = explode,
        colors = color,
        autopct = "%0.2f%%", #Shows value in percentage.
        shadow = True, #Gives Drop Shadow to Pie Chart.
        radius = 1.4, #Increase Size of PIE CHART,
        startangle = 90, #Rotates chart.
        textprops = textprops, #change properties of text.
       
       )
plt.legend(loc = 1)
plt.show()
```

## How to SAVE FIGURE?

```python
# For help press "SHIFT + TAB"  when in between paranthesis
# and click the arrow button to know more about the function.



plt.pie(class1_students,
        labels = classes,
        explode = explode,
        colors = color,
        autopct = "%0.2f%%", #Shows value in percentage.
        #shadow = True, #Gives Drop Shadow to Pie Chart.
        radius = 1.4, #Increase Size of PIE CHART,
        startangle = 90, #Rotates chart.
        textprops = textprops, #change properties of text.
       
       )
plt.legend(loc = 1)

plt.savefig("pie_chart",
           dpi = 100)

plt.show()
```
