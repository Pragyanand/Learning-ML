```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sklearn, scipy
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

```python
data.columns
```

```python
data.loc[data["Sex"] == "male", "Sex"]  = 0
data.loc[data["Sex"] == "female", "Sex"]  = 1
data.describe()
```

```python
data.info()
```

## Plot the histogram

```python
data.Survived.value_counts(normalize = True).plot(kind='bar')
plt.title("Graph Depicting Survivor Percentage")
```

```python
plt.scatter(data.Survived, data.Age, alpha = 0.1)
plt.title("Age vs Survived")
```

## Find the Correlation

```python
coor_matrix = data.corr()
coor_matrix['Survived'].sort_values(ascending = True)
```

```python
from pandas.plotting import scatter_matrix 
attributes = ["Pclass", "Fare","Parch", "Age", "Survived"]
scatter_matrix(data[attributes], figsize = (20,16))
```

## Filling the Missing Values

```python
data["Age"] = data["Age"].fillna(data["Age"].dropna().median())
```

## Changing Categorical Information to Numbers

```python
#data.loc[data["Sex"] == "male", "Sex"]  = 0
#data.loc[data["Sex"] == "female", "Sex"]  = 1
```

```python
target = data["Survived"].values
features = data[["Pclass", "Fare","Parch", "Age"]].values
```

## Selecting the Algorithm

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

#model = LogisticRegression()
#model = RandomForestClassifier(random_state = 1)
model = DecisionTreeClassifier(random_state = 1)

a = model.fit(features, target)

scores = cross_val_score(model, features, target, scoring = "accuracy", cv=50)
print(scores)
print("\n Mean of Score is:" , scores.mean())

```

```python
print (a.score(features, target))
```
